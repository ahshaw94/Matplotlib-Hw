```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merged_df = pd.merge(ride_df,city_df,on=["city"], how="left")
# Display the data table for preview
merged_df.head()
```

```python
#grouping by city to showcase relationship between variables
ride_city = merged_df.groupby("city")

avg_fare = ride_city["fare"].mean()

rides_per_city = ride_city["ride_id"].count()

drivers_per_city = ride_city["driver_count"].mean()

city_type = ride_city["type"].unique()

#create df to hold new variables

city_df = pd.DataFrame({"Average Fare":avg_fare,
                       "Total Rides/City":rides_per_city,
                        "Total Drivers/City":drivers_per_city,
                        "City Type":city_type})

city_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban_type = city_df[city_df["City Type"] == "Urban"]
suburban_type = city_df[city_df["City Type"] == "Suburban"]
rural_type = city_df[city_df["City Type"] == "Rural"]

# Build the scatter plots for each city types

plt.scatter(urban_type["Total Rides/City"],urban_type["Average Fare"],s=urban_type["Total Drivers/City"]*5,color="coral",edgecolor="black",label="Urban",alpha=.75)
plt.scatter(suburban_type["Total Rides/City"],suburban_type["Average Fare"],s=suburban_type["Total Drivers/City"]*5,color="skyblue",edgecolor="black",label="Suburban",alpha=.75)
plt.scatter(rural_type["Total Rides/City"],rural_type["Average Fare"],s=rural_type["Total Drivers/City"]*5,color="gold",edgecolor="black",label="Rural",alpha=.75)

# Incorporate the other graph properties

plt.xlabel("Total Number of Rides/City")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid()

# Create a legend

legend=plt.legend(loc="best")

# Incorporate a text label regarding circle size

plt.text(45,35,"Note:Circle Size Increases based on Driver Count per City")

# Save Figure

plt.savefig("Images/PyberData.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents

fare_type = merged_df.groupby("type")

total_fares = fare_type["fare"].sum()

labels=total_fares.index

colors=["gold","skyblue","coral"]

explode=[0.0,0.0,0.1]

# Build Pie Chart

plt.pie(total_fares,labels=labels,shadow=True,colors=colors,explode=explode,startangle=140,autopct="%1.2f%%")

plt.axis("equal")
plt.title("% of Total Fares by City Type")

plt.show()

# Save Figure
plt.savefig("Images/FaresData.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_type = merged_df.groupby("type")

total_rides = ride_type["ride_id"].count()

labels=total_rides.index

colors=["gold","skyblue","coral"]

explode=[0.0,0.0,0.1]

# Build Pie Chart

plt.pie(total_rides,labels=labels,shadow=True,colors=colors,explode=explode,startangle=140,autopct="%1.2f%%")

plt.axis("equal")
plt.title("% of Total Rides by City Type")

plt.show()

# Save Figure
plt.savefig("Images/RidesData.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_type = merged_df.groupby("type")

total_drivers = driver_type["driver_count"].sum()

labels=total_drivers.index

colors=["gold","skyblue","coral"]

explode=[0.0,0.0,0.1]

# Build Pie Chart

plt.pie(total_drivers,labels=labels,shadow=True,colors=colors,explode=explode,startangle=140,autopct="%1.1f%%")

plt.axis("equal")
plt.title("% of Total Drivers by City Type")

plt.show()

# Save Figure
plt.savefig("Images/DriversData.png")
```
